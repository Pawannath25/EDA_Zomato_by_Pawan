## Importing Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv(r"C:\Elevate lab internship\zomato.csv")
df
```

```python
import numpy as np
```

```python
# Initial checks
print(df.info())  # Check data types and missing values
print(df.describe())  # Summary stats for numerical columns
print(df.nunique())  # Unique values per column
```

```python
df.isnull().sum()
```

## Data Cleaning

```python
df.drop(columns=["Unnamed: 0.1","Unnamed: 0"],inplace=True)
```

```python
df
```

```python
# Handle missing values
df.dropna(subset=['rate (out of 5)'], inplace=True)
```

```python
# Group by restaurant type and calculate median cost for each group
group_median = df.groupby('restaurant type')['avg cost (two people)'].transform('median')

# Fill nulls with group-specific medians
df['avg cost (two people)'] = df['avg cost (two people)'].fillna(group_median)
```

```python
# Convert categorical features to binary
df['online_order'] = df['online_order'].map({'Yes': 1, 'No': 0})
df['table booking'] = df['table booking'].map({'Yes': 1, 'No': 0})
```

```python
# Split cuisines into lists
df['cuisines type'] = df['cuisines type'].str.split(', ')
```

## Correlation Heatmap

## Price-to-Rating Ratio: Identify Value-for-Money Restaurants

```python
# Create a new metric: Rating per ₹100 spent
df['rating_per_100'] = (df['rate (out of 5)'] / df['avg cost (two people)']) * 100

# Top 10 restaurants with the best value
top_value = df.sort_values('rating_per_100', ascending=False).head(10)[['restaurant name', 'rating_per_100', 'area']]
print(top_value)
```

"Srinidhi Sagar Food Line and Deluxe offer exceptional value, delivering the highest rating-per-100-rupees ratio, making them standout budget-friendly dining options."

"Banashankari emerges as a hotspot for value-for-money eateries, with three restaurants in the top 10 list, catering to cost-conscious diners without compromising quality."

## Cuisine Combination Analysis

```python
from itertools import combinations
from collections import Counter

# Extract all cuisine pairs
cuisine_pairs = df['cuisines type'].dropna().apply(lambda x: list(combinations(x, 2)))
all_pairs = [pair for sublist in cuisine_pairs for pair in sublist]
pair_counts = Counter(all_pairs).most_common(10)

# Plot top pairs
sns.barplot(x=[count[1] for count in pair_counts], y=[str(count[0]) for count in pair_counts])
plt.title('Top Cuisine Combinations')
```

 OBSERVATION
"North Indian-Chinese fusion dominates as the most popular cuisine pairing, reflecting India's love for this classic cross-cultural flavor combination."

"The frequent appearance of Biryani in top combinations highlights its versatility as both a standalone dish and a complementary element in multi-cuisine dining experiences."

## Premium V/S Budget Restaurant Amalysis

```python
# Segment restaurants into premium (top 25% cost) and budget
premium_threshold = df['avg cost (two people)'].quantile(0.75)
df['segment'] = np.where(df['avg cost (two people)'] > premium_threshold, 'Premium', 'Budget')

# Compare ratings and online orders
sns.boxplot(x='segment', y='rate (out of 5)', hue='online_order', data=df)
plt.title('Premium vs. Budget: Ratings & Online Orders')
```

"Premium restaurants maintain higher ratings than budget options regardless of online order availability, confirming that perceived quality justifies higher prices."

"The wider rating gap between online/non-online budget restaurants suggests digital ordering creates disproportionately greater value for affordable dining experiences."

## Geospatial Analysis(Area-Wise-Insights)

```python
# Create a pivot table for area vs. online orders
area_online = pd.pivot_table(df, index='area', columns='online_order', values='restaurant name', aggfunc='count').fillna(0)
area_online['online_order_ratio'] = area_online[1] / (area_online[0] + area_online[1])

# Top areas with low online adoption
low_online_areas = area_online[area_online['online_order_ratio'] < 0.5].index.tolist()[:5]
print("Areas needing online push:", low_online_areas)
```

"Traditional food hubs like Basavanagudi and Brigade Road show surprisingly low online order adoption (<50%), revealing untapped potential for digital expansion in these established dining areas."

"The presence of tech-centric zones like Electronic City and HSR in low-adoption areas suggests a mismatch between technological infrastructure and restaurant digitalization efforts."

## Operational Efficiency: Online Orders + Table Booking

```python
# Compare ratings for restaurants offering both services
df['both_services'] = (df['online_order'] == 1) & (df['table booking'] == 1)
sns.violinplot(x='both_services', y='rate (out of 5)', data=df)
plt.title('Ratings for Restaurants Offering Both Services')
```

"Restaurants offering both online orders and table booking (True) show visibly higher average ratings than those without both services, highlighting the cumulative benefit of convenience features."

"The clear performance gap suggests customers increasingly expect and reward restaurants that provide multiple service options, making this a key competitive differentiator."

## Time-Based Demand Prediction (Proxy Analysis)

```python
# Bin restaurants by number of ratings (proxy for longevity)
df['popularity_bin'] = pd.qcut(df['num of ratings'], q=3, labels=['Low', 'Medium', 'High'])

# Compare average cost and ratings across bins
sns.scatterplot(x='avg cost (two people)', y='rate (out of 5)', hue='popularity_bin', data=df)
plt.title('Cost vs. Ratings by Popularity')
```

"High-popularity restaurants maintain strong ratings (4.0+) across all price points, proving that customer volume doesn't compromise perceived quality when properly managed."

"The steepest rating decline occurs in low-popularity establishments as prices exceed ₹3000, suggesting unknown restaurants face tougher scrutiny at higher price levels."

## Cuisine Saturation Analysis

```python
# Calculate cuisine count vs. average rating
cuisine_stats = df.explode('cuisines type').groupby('cuisines type').agg(
    count=('cuisines type', 'count'),
    avg_rating=('rate (out of 5)', 'mean')
).reset_index()

# Plot saturation vs. rating
sns.scatterplot(x='count', y='avg_rating', data=cuisine_stats, hue='count', size='count')
plt.title('Cuisine Saturation vs. Rating')
```

"The inverse relationship between cuisine saturation and ratings suggests that as a cuisine type becomes more common (higher count), its average rating tends to decrease slightly, indicating potential quality dilution in oversaturated markets."

"The most highly-rated cuisines (4.0-4.5) appear in the mid-range of saturation (500-1500 count), suggesting an optimal balance between availability and quality maintenance."

## Statistical Validation with Hypothesis Testing

```python
from scipy.stats import ttest_ind

# Test if online orders have significantly higher ratings
online = df[df['online_order'] == 1]['rate (out of 5)']
offline = df[df['online_order'] == 0]['rate (out of 5)']
t_stat, p_value = ttest_ind(online, offline)
print(f"p-value: {p_value:.4f}")  # If p < 0.05, difference is significant
```

"The extremely low p-value (≈0.0000) confirms that restaurants offering online orders receive statistically higher ratings than those without this service."

"This significant finding suggests that digital ordering capability correlates with better customer satisfaction, possibly due to convenience or operational efficiency."

## Competitor Benchmarking/

```python
# Compare chains vs. independent restaurants
chains = df['restaurant name'].value_counts()[df['restaurant name'].value_counts() > 3].index.tolist()
df['is_chain'] = df['restaurant name'].isin(chains)

# Compare ratings and costs
print(df.groupby('is_chain')[['rate (out of 5)', 'avg cost (two people)']].mean())
```

"Independent restaurants (non-chains) maintain a competitive 3.51 average rating, proving standalone eateries can match chain establishments in quality."

"With an average meal cost of ₹542 for two, independent restaurants offer mid-range pricing, suggesting they strike a balance between affordability and quality in the dining market."
